##1 .Write a for loop program to print all elements of a tuple.


```python
a = (1,2,3,4,5)

for i in a:
  print(i)
```

##2. Write a while loop program to print all elements of a tuple using indexes.


```python
a = (1,2,3,4,5)
i = 0

while i < len(a):
  print(a[i])
  i = i+1
```

##3. Write a for loop program to count how many even numbers are present in a tuple.

```python
a = (1,2,3,4,5)
c = 0
for i in a:
  if i % 2 == 0:
    c = c + 1
print('count')
```

##4. Write a while loop program to find the largest value in a tuple.


```python
a = (1,22,3,4,61)
large = a[0]
i = 0

while i < len(a):
  if a[i] > large:
    large = a[i]
  i = i + 1
print(large)
```

##5. Write a for loop program to print the index and value of each tuple element.

##6. Write a while loop program to reverse a tuple manually without using slicing.


```python
a=(1,2,3,4,5)
n = len(a)-1

while n>=0:
  print(a[n],end=' ')
  n=n-1

```

##7. Write a for loop program to create a new tuple containing only odd numbers from an existing tuple.


```python
a = (1,2,3,4,5)
b = []

for i in a:
  if i%2!=0:
   b.append(i)
print(tuple(b))

```

##10. Write a while loop program to count how many times a particular element occurs in a tuple.

##11. Write a for loop program to find how many strings in a tuple have length greater than 5.


```python
a=('ravali','shiv','shivansh','andeep')
c=0
i=0
while i<len(a):
  if len(a[i])>5:
    c=c+1
  i=i+1
print(f'{c} strings have length greater than 5')
```

##13. Write a for loop program to separate elements of a tuple into two new tuples: one with even numbers and one with odd numbers.

```python
a=(10,11,12,13,14,15,16)
o = []
e = []
i = 0
while i <len(a):
  if a[i]%2==0:
    e.append(a[i])
  else:
    o.append(a[i])
  i=i+1
odd=tuple(o)
even=tuple(e)
print(even)
print(odd)
```

##14. Write a while loop program to copy elements from one tuple to a new tuple.
