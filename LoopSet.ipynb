##1. Write a for loop program to print each element of a set.

```python
a = {1,2,3,4,5}
b = list(a)
print(type(b))
i = 0
while i<len(b):
  print(b[i])
  i = i+1
```

##3. Write a for loop program to count how many even numbers are in a set.


```python
a = {1,2,3,4,5,6}
count = 0

for i in a:
  if i % 2==0:
    count = count+1
print(f'{count} even numbers')
```

##4. Write a while loop program to find the largest element in a set.


```python
a = {1,2,3,4,5,6}
b = list(a)
i = 1
l=b[i]


while i < len(b):
  if b[i]>l:
    l=b[i]
  i = i+1
print(f'{l} large number')


```

##14. Write a while loop program to print elements of a set in reverse order after converting to a list.


```python
a = {12,13,14,16}
b=list(a)
n = len(b)-1
while n>=0:
  print(b[n])
  n=n-1
```
