##1. Write a for loop program to print each element of a list on a new line.


```python
a = [1,2,3,4]


for i in a:
  print(i)
```

##2. Write a while loop program to print all items of a list using indexing.


```python
a = [20,3,36,4]

i = 0

while i < len(a):
  print(a[i])
  i = i + 1
```

##3. Write a for loop program to find the sum of all elements in a list.

```python
a = [1,2,3,4]
sum = 0


for i in a:
  sum = sum + i
print(sum)
```

##4. Write a while loop program to find the largest number in a list.


```python
a = [14,3,4,6]

large = a[0]

while i < len(a):
  if a[i]>large:
    large = a[i]
    i = i +1
print(large)
```

##6. Write a while loop program that creates a new list containing only odd numbers from the original list.


```python
a = [10,20,30,40,5,6,8,9]
b = []

for i in a:
  if i%2!=0:
     b.append(i)

print(b)
```

##7. Write a for loop program to count how many times a specific element appears in a list.

```python
a = [1,2,3,1,2,2,4,5,2,8]
c = 0

for i in a:
  if i == 1:
    c = c + 1
print(c)
```

##8. Write a while loop program to reverse a list without using the reverse function.


```python
a = [1,2,3,4,5]
i = len(a)
b = []
while i > 0:
  b.append(i)
  i = i-1
print(b)
```

##12. Write a while loop program that stops when it encounters the number 0 in a list.


```python
a = [1,4,6,0,8,9]
i = 0
while i<len(a):
  if a[i] == 0:
    break
  print(a[i])
  i = i+1
```

##14. Write a while loop program to count how many integers in the list are divisible by 5.


```python
a = [10,12,15,20,25,26]
i = 0
c = 0
while i<len(a):
  if a[i]%5 == 0:
    c = c+1
  i=i+1
print(f'{c}')
```

##15. Write a for loop program to separate elements into two new lists: one with even numbers and one with odd numbers.


```python
a = [12,3,4,5,6]
b = []
c = []
for i in a:
  if i%2 == 0:
    b.append(i)
  else:
    c.append(i)

print(b)
print(c)
```
