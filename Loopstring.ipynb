##8. Write a while loop program that stops printing characters when it encounters the letter a.


```python
a = 'Shivansh'
i = 0
while i<len(a):
  if a[i]=='a':
    break
  print(a[i])
  i=i+1
```

##15. Write a for loop program to count how many alphabetic characters, digits, and special characters are in a string.


```python
a='ravali@123#'
al = 0
digits = 0
special = 0



for i in a:
  if i.isalpha():
    al = al + 1
  elif i.isnumeric():
    digits = digits + 1
  else:
    special = special + 1
print(f'alphabets:{al}  digits:{digits}  special:{special}')
```
